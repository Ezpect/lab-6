```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
  df = pd.read_csv('Balloon Race Data Breaches - LATEST - breaches.csv')
  print(df.head())
except FileNotFoundError:
  print("Error: File 'Balloon Race Data Breaches - LATEST - breaches.csv' not found.")

```

```python
df.info()

df.describe(include='all')
```

```python
keywords = ['graphics', 'GPU', 'rendering', 'visualization', 'computer graphics', 'hacker', 'data', 'breaches']

graphics_df = df[df['story'].str.contains('|'.join(keywords), case=False, na=False) |
                 df['organisation'].str.contains('|'.join(keywords), case=False, na=False)]

graphics_df.head(10)
```

```python

graphics_df.describe(include='all')

graphics_df.isnull().sum()

graphics_df['method'].value_counts()
```

```python
breaches_over_time = graphics_df.groupby('date').size()

common_methods = graphics_df['method'].value_counts()

data_types = graphics_df['data sensitivity'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
breaches_over_time.plot(kind='line')
plt.title('Number of Breaches Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Breaches')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
common_methods.plot(kind='bar')
plt.title('Distribution of Breach Methods')
plt.xlabel('Method of Breach')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

data_types.plot(kind='pie', autopct='%1.1f%%', wedgeprops=dict(width=0.6))

plt.title('Types of Data Compromised')
plt.ylabel('')
plt.show()
```
